#### Write a simple program that reads a "txt" file and removes all the stop words.

```python
import csv
import nltk
import string 
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def wordcount(textfile, countfile):
    w_c = []
    dictionary = { }
    with open(textfile,'r') as readfile:
        with open(countfile, 'w+') as csv_file:
            for text in readfile:
                word = text.split(" ")
                for element in word:
                    if element.lower() not in stop_words:
                        w_c.append(element)
            for w in w_c:
                if w in dictionary.keys():
                    count = dictionary[w]
                    count = count + 1
                    dictionary[w] = count
                else:
                    dictionary[w] = 1  
        return dictionary                
def createcsv(dec):
    with open(countfile, 'w') as csv_file:
        writecount = csv.writer(csv_file)
        for key, value in dec.items():
            writecount.writerow([key, value])

textfile ='C:/Users/Yuri/Desktop/churchill.txt'
countfile ='C:/Users/Yuri/Desktop/church-wordcount9.csv'
final = wordcount(textfile, countfile)
createcsv(final)
```
